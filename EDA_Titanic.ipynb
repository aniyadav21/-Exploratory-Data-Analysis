# EDA - Titanic Dataset

This notebook performs exploratory data analysis (EDA) on the Titanic dataset.
Tools: Pandas, Seaborn, Matplotlib.


```python
import pandas as pd
df = pd.read_csv('train.csv')
df.head()
```

```python
# Basic info
print(df.shape)
print(df.info())
print(df.isnull().sum())
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()

# Survived count
sns.countplot(x='Survived', data=df)
plt.show()

# Survival by sex
sns.countplot(x='Sex', hue='Survived', data=df)
plt.show()

# Survival rate by Pclass
pclass_rate = df.groupby('Pclass')['Survived'].mean().reset_index()
sns.barplot(x='Pclass', y='Survived', data=pclass_rate)
plt.show()

# Age distribution
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.show()

# Fare distribution
sns.histplot(df['Fare'].dropna(), bins=30, kde=True)
plt.show()

# Correlation heatmap
num_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
sns.heatmap(df[num_cols].corr(), annot=True, fmt='.2f')
plt.show()
```

## Observations
1. Females have higher survival rates than males.
2. First-class passengers have higher survival rates.
3. Age and Fare distributions indicate skewness; consider log transforms for Fare.
